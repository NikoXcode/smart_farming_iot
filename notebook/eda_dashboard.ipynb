```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

current_dir = os.getcwd() 
project_root = os.path.dirname(current_dir)
clean_data_path = os.path.join(project_root, 'data', 'processed', 'cleaned_farm_data.csv')

df = pd.read_csv(clean_data_path)
sns.set_theme(style="whitegrid")
print("Data loaded successfully!")
```

```python
plt.figure(figsize=(10, 5))
sns.countplot(data=df, x='crop_type', hue='crop_type', palette='viridis', legend=False)
plt.title('Crop Distribution Across Farms', fontsize=14, fontweight='bold')
plt.xlabel('Crop Type', fontsize=12)
plt.ylabel('Number of Farms', fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.scatterplot(data=df, x='temperature_C', y='soil_moisture_%', hue='crop_type', palette='deep', alpha=0.7)
plt.title('Temperature vs. Soil Moisture by Crop Type', fontsize=14, fontweight='bold')
plt.xlabel('Ambient Temperature (°C)', fontsize=12)
plt.ylabel('Soil Moisture (%)', fontsize=12)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
